```python
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    'Product': [f'Product_{i}' for i in range(1, 21)],
    'Price': np.random.uniform(5.0, 200.0, 20).round(2), # Prices between $5 and $200
    'Stock': np.random.randint(0, 150, 20) # Stock quantities between 0 and 150
}
df = pd.DataFrame(data)

# Display the DataFrame in console format
print('Sample Product Data:')
print(df.to_string())

# Display the first 5 rows of the DataFrame in GUI format (Colab's rich output)
print('\nFirst 5 rows of the DataFrame:')
display(df.head())
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot
plt.figure(figsize=(10, 6)) # Set the figure size for better readability
plt.scatter(df['Stock'], df['Price'], alpha=0.7, color='skyblue', edgecolors='w', s=100) # Create the scatter plot

# Add titles and labels
plt.title('Relationship Between Stock and Price of Products') # Title of the plot
plt.xlabel('Stock Quantity') # X-axis label
plt.ylabel('Price ($)') # Y-axis label

# Add grid for better readability
plt.grid(True, linestyle='--', alpha=0.6)

# Add annotations for a few products to make it more informative (optional)
for i in range(len(df)):
    if i % 5 == 0: # Annotate every 5th product for clarity
        plt.annotate(df['Product'][i], (df['Stock'][i], df['Price'][i]),
                     textcoords="offset points", xytext=(0,10),
                     ha='center', fontsize=8)

# Display the plot
plt.show() # This will display the plot in GUI format
```
